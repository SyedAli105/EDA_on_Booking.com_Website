# Getting hotel names and info using Web Scrapping:

```python
hotels_info={'Hotel_Name':[],'Description':[],'Rating':[]}
for i in range(25):
    hotels_info['Hotel_Name'].append(soup.find_all('h3')[i].text.replace('Opens in new window',''))
    hotels_info['Description'].append(soup.find_all('div',{'class':'d8eab2cf7f'})[i].text)
    hotels_info['Rating'].append(soup.find_all('div',{'class':'b5cd09854e d10a6220b4'})[i].text)
```

```python
hotels_data=pd.DataFrame(hotels_info)
hotels_data.to_csv('C:\\Users\\Syed Muhammad Ali\\Desktop\\hotel.csv')
hotels_data
```
